```python
from nltk import ngrams

sentence = "Hey this is Ananya"

n = 1

n_grams = ngrams(sentence.split(), n)
for i in n_grams:
    print(i)
```

```python
from nltk import everygrams

line = "Let us understand everygrams with a simple example below"

l = line.split()

list(everygrams(l))
```

```python
sentence2 = "This code will give us bigrams"
n = 2
n_grams = ngrams(sentence2.split(), n)
for j in n_grams:
    print(j)
```

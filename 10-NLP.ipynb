```python
import pandas as pd
import math
import numpy as np
```

```python
corpus = ['It is going to rain today' , 'Today I am not going outside because I think it might rain', 'I am going to watch the rains this season']
words_set = set()
for aa in corpus:
    words = aa.split(' ')
    words_set = words_set.union(set(words))
print('Number of words in the corpus:',len(words_set))
print('The words in the corpus: \n', words_set)
```

```python
n_docs = len(corpus)         #·Number of documents in the corpus
n_words_set = len(words_set) #·Number of unique words in the 

df_tf = pd.DataFrame(np.zeros((n_docs, n_words_set)), columns=words_set)

# Compute Term Frequency (TF)
for ij in range(n_docs):
    words = corpus[ij].split(' ') # Words in the document
    for ww in words:
        df_tf[ww][ij] = df_tf[ww][ij] + (1 / len(words))
        
df_tf

#def computeTF(words_set, corpus):
#    tfDict = {}
#    corpusCount = len(df_tf)
#    for word, count in df_tf.items():
#        tfDict[word] = count/float(corpusCount)
#    return(tfDict)
#running our sentences through the tf function:
#tf = computeTF(words_set, corpus)
#Converting to dataframe for visualization
#tf = pd.DataFrame(tf)
#print(tf)
```

```python
print("IDF of: ")
invdf = {}
for all in words_set:
    kk = 0    # number of documents in the corpus that contain this word
    for z in range(n_docs):
        if all in corpus[z].split():
            kk += 1
    invdf[all] =  np.log10(n_docs / kk)
    print(f'{all:>15}: {invdf[all]:>10}' )
```

```python
df_tf_idf = df_tf.copy()

for w in words_set:
    for i in range(n_docs):
        df_tf_idf[w][i] = df_tf[w][i] * invdf[w]
        
df_tf_idf
```

```python
import nltk
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import drive
import os
drive.mount('/content/drive')
```

```python
# Mount Google Drive to access the text files

# Define directory path of text files
dir_path = '/content/drive/MyDrive/ALL-SEM-08/Nlp-lab/stories/'
enc = 'iso-8859-15'
# Initialize a list to store the contents of all text files
corpus = []

# Loop through all the text files in the directory
for file_name in os.listdir(dir_path):
    if file_name.endswith('.txt'):
        # Open each file and read its contents
        with open(os.path.join(dir_path, file_name), 'r', encoding=enc) as file:
            content = file.read()
            # Append the contents of the file to the corpus list
            corpus.append(content)
```

```python
# initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
```

```python
# fit and transform the corpus using the vectorizer
tfidf = tfidf_vectorizer.fit_transform(corpus)
```

```python
# define the question
question = "What was the name of the flute?"
# transform the question using the vectorizer
question_tfidf = tfidf_vectorizer.transform([question])
```

```python
# calculate the cosine similarity between the question and each document in the corpus
cosine_similarities = cosine_similarity(question_tfidf, tfidf).flatten()
```

```python
# get the index of the document with the highest cosine similarity
best_index = np.argmax(cosine_similarities)
# print the answer
print(corpus[best_index])
```

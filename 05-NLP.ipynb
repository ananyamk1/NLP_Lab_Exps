```python
from nltk.tokenize import word_tokenize
from nltk.tag import DefaultTagger
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
text = """Today morning, we are performing nlp experiment 4 named part of speech tagging"""

tokens = nltk.word_tokenize(text)
```

```python
tagged = nltk.pos_tag(tokens)
print(tagged)
```

```python
text2 = DefaultTagger('JJ')
text2.tag(['slowly', 'gracefullly', 'assertively'])
```

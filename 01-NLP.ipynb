```python
htext = """Wikipedia is a multilingual free online encyclopedia written and maintained by a community of volunteers, known as Wikipedians, through open collaboration and using a wiki-based editing system."""
htext
```

```python
from nltk.tokenize import sent_tokenize
```

```python
nltk.download('punkt')
```

```python
sent_tokenize(htext)
```

```python
from nltk.tokenize import word_tokenize
word_tokenize(htext)
```

```python
from nltk.tokenize import TreebankWordTokenizer
htoken = TreebankWordTokenizer()
htoken.tokenize(htext)
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(htext)
```

```python
import nltk
import numpy as np
```

```python
states = ['noun', 'verb', 'adjective']
observations = ['the', 'bird', 'is', 'beautiful']
```

```python
initial_prob = {'noun': 0.6, 'verb': 0.1, 'adjective': 0.3}
```

```python
transition_prob = {'noun': {'noun': 0.4, 'verb': 0.3, 'adjective': 0.3},
                   'verb': {'noun': 0.2, 'verb': 0.5, 'adjective': 0.3},
                   'adjective': {'noun': 0.1, 'verb': 0.3, 'adjective': 0.6}}
```

```python
emission_prob = {'noun': {'the': 0.5, 'bird': 0.3, 'is': 0.2},
                 'verb': {'the': 0.1, 'bird': 0.4, 'is': 0.5},
                 'adjective': {'the': 0.2, 'bird': 0.1, 'is': 0.6, 'beautiful': 0.3}}
```

```python
emission_matrix = np.zeros((len(states), len(observations)))
for i in range(len(states)):
    for j in range(len(observations)):
        emission_matrix[i][j] = emission_prob[states[i]].get(observations[j], 0)
```

```python
# calculate the transition matrix
transition_matrix = np.zeros((len(states), len(states)))
for i in range(len(states)):
    for j in range(len(states)):
        transition_matrix[i][j] = transition_prob[states[i]].get(states[j], 0)
```

```python
# normalize the matrices
emission_matrix /= emission_matrix.sum(axis=0, keepdims=True)
transition_matrix /= transition_matrix.sum(axis=0, keepdims=True)
```

```python
# print the matrices
print('Emission Matrix:', emission_matrix)
print('Transition Matrix:', transition_matrix)
```

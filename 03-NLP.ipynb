```python
# Text: The original word text.
# Lemma: The base form of the word.
# POS: The simple UPOS part-of-speech tag.
# Tag: The detailed part-of-speech tag.
# Dep: Syntactic dependency, i.e. the relation between tokens.
# Shape: The word shape â€“ capitalization, punctuation, digits.
# is alpha: Is the token an alpha character?
# is stop: Is the token part of a stop list, i.e. the most common words of the language?
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
say = nlp("Do not judge each day by the harvest you reap but by the seeds that you plant")
```

```python
for token in say:
    print(f"Word: {token.text}\nRoot: {token.lemma_}\nPOS: {token.pos_}\nTAG: {token.tag_}\nEXPLANATION: {spacy.explain(token.tag_)}\nDEP: {token.dep_}\nSHAPE: {token.shape_}\nIS ALPHA: {token.is_alpha}\nIS STOPWORD: {token.is_stop}\n\n")
```

```python
# import libraries
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
# download nltk corpus (first time only)
import nltk
#nltk.download('all')
```

```python
# create preprocess_text function
text1 = "This food is delicious and the best thing I have eaten."
text2 = "this pizza was great"
text3 = "the burger is bad"
text4 = "the dish I ordered is a pie"
text5 = "this serving looks so good"

def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(text.lower())
    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]
    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    # Join the tokens back into a string
    processed_text = ' '.join(lemmatized_tokens)
    print(processed_text)
    return 

preprocess_text(text1)
preprocess_text(text2)
preprocess_text(text3)
preprocess_text(text4)
preprocess_text(text5)
```

```python
# initialize NLTK sentiment analyzer
analyzer = SentimentIntensityAnalyzer()
# create get_sentiment function
def get_sentiment(text):
    scores = analyzer.polarity_scores(text)
    if scores['pos'] > 0.5:
      sentiment = "positive"
    elif scores['neg'] > 0.5:
      sentiment = "negative"
    elif scores['neu'] > 0.4:
      sentiment = "neutral"
    print(sentiment)
    return sentiment
    
# apply get_sentiment function
rev1 = get_sentiment(text1)
rev2 = get_sentiment(text2)
rev3 = get_sentiment(text3)
rev4 = get_sentiment(text4)
rev5 = get_sentiment(text5)
```
